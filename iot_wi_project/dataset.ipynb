```python
import pandas as pd
```

```python
file_path = '/Users/chiaradraghini/Documents/iot_wi_project/data/capture/'

data = pd.read_csv(file_path+'project_iotwi_capture.csv')
```

```python
data_filtered_protocol = data[data['Protocol'] == "UDP"]
data_filtered_protocol
```

```python
cols_to_keep = ['Time', 'Source', 'Destination', 'Length','Delta']
data_filtered = data_filtered_protocol.loc[:,cols_to_keep]
data_filtered
```

```python
lst = data_filtered["Time"].to_list()
array = []
array.append(lst)
```

```python
data_filtered2 = data_filtered.set_index("Time")
data_filtered2
```

```python
x = 0
y = 0.5
rows = []
array_source = []
while (y <= 602.849661):
    for i in range(len(array)):
        
        df_iteration = data_filtered[data_filtered['Time'].between(x, y)]
        df_iteration_pc = df_iteration[df_iteration['Source'] == "172.20.10.2"]
        df_iteration_ph = df_iteration[df_iteration['Source'] == "144.195.65.24"]

        mean_l = df_iteration['Length'].mean()
        std_l = df_iteration['Length'].std()
        median_l = df_iteration['Length'].median()
        mean_d = df_iteration['Delta'].mean()
        std_d = df_iteration['Delta'].std()
        median_d = df_iteration['Delta'].median()
        
        inb = len(df_iteration_ph)
        outb = len(df_iteration_pc)
        tot_pkts = len(df_iteration)        
        
        rows.append([(x,y), mean_l, std_l, median_l, mean_d, std_d, median_d, inb, outb, tot_pkts])
        
    x = y
    y = y + 0.5
#print(df_iteration)
#print(mean)
    
    df = pd.DataFrame(rows)
df.columns = ["Time Interval","Average Packet Size","Standard Deviation Packet Size","Median Packet Size","Average Packet Intervals","Standard Deviation Packet Intervals", "Median Packet Intervals" ,"Inbound Packets", "Outbound Packets","Total Packets"]
df
```
